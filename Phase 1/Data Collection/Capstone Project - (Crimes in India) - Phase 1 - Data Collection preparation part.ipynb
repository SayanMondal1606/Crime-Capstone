THE PROVIDED DATA FOR THE CAPSTONE PROJECT CONTAINS DATA OF THE YEAR 2012 ALSO SO TO FIND THE POPULATION OF DATA IN THE YEAR 2012 THE DATA WAS WEB SCRAPPED USING BEAUTIFUL SOUP FROM THE WEBSITE : "https://statistics.py.gov.in/state-wise-population-2012?page=3#:~:text=State%2Dwise%20Population%20%2D%202012%20;" TO ENSURE PROPER DATA COLLECTION.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://statistics.py.gov.in/state-wise-population-2012?page=3#:~:text=State%2Dwise%20Population%20%2D%202012%20;'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', class_='table table-bordered table-hover')
rows = table.find_all('tr')

data = []
for row in rows:
    columns = row.find_all('td')
    row_data = [column.text.strip() for column in columns]
    data.append(row_data)

# Create a DataFrame from the scraped data
df = pd.DataFrame(data, columns=['State Code', 'State', 'Total Population', 'Male Population', 'Female Population', 'Percentage', 'Sex_Ratio'])

# Print the DataFrame
print(df)
```

REQ. NUMBER - 2

Literacy Rate in each state

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.census2011.co.in/literacy.php'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table')
rows = table.find_all('tr')

data = []

for row in rows[1:]: # Skip first row as it contains header
    columns = row.find_all('td')
    state = columns[1].text
    literacy_rate = columns[2].text
    Male= columns[3].text
    Female = columns[4].text
    Percentage_Change = columns[5].text
    
    data.append([state, literacy_rate, Male, Female, Percentage_Change])

# Create DataFrame
literacy_rate = pd.DataFrame(data, columns=['State', 'Literacy Rate', 'Male Literacy', 'Female Literacy', 'Percentage Change'])

# Display the DataFrame
print(literacy_rate)
```

REQ. NUMBER - 3

Area of each state

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_area'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', class_='wikitable')
rows = table.find_all('tr')

data = []
for row in rows:
    columns = row.find_all('td')
    row_data = []
    for column in columns:
        row_data.append(column.text.strip())
    data.append(row_data)

df = pd.DataFrame(data)
print(df)
```
